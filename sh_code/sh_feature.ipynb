```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# mean 계산
df['mean'] = signal_data.mean(axis=1)

# absolute mean 계산 (절대값의 평균)
df['absolute_mean'] = signal_data.abs().mean(axis=1)

# absolute max 계산 (절대값의 최대값)
df['absolute_max'] = signal_data.abs().max(axis=1)

# std 계산 (표준편차)
df['std'] = signal_data.std(axis=1)

# variance 계산 (분산)
df['variance'] = signal_data.var(axis=1)

# peak-to-peak 계산 (최대값과 최소값의 차이)
df['peak_to_peak'] = signal_data.max(axis=1) - signal_data.min(axis=1)

# peak 계산 (최대값)
df['peak'] = signal_data.max(axis=1)

# skewness 계산 (왜도)
df['skewness'] = signal_data.apply(lambda row: skew(row), axis=1)

# kurtosis 계산 (첨도)
df['kurtosis'] = signal_data.apply(lambda row: kurtosis(row), axis=1)

# RMS 계산 (Root Mean Square, 제곱 평균 근)
df['rms'] = np.sqrt((signal_data ** 2).mean(axis=1))

# crest factor 계산 (최대값을 RMS로 나눈 값)
df['crest_factor'] = df['peak'] / df['rms']

# impulse factor 계산 (절대 최대값을 절대 평균값으로 나눈 값)
df['impulse_factor'] = df['absolute_max'] / df['absolute_mean']

# 고차원 특성 추가
df['margin'] = df['peak'] - df['mean']  # Margin: 평균값
df['power'] = df['rms'] ** 2  # Power: RMS^2
df['shape_factor'] = df['rms'] / df['absolute_mean']  # Shape factor: RMS / 절대 평균
df['clearance_factor'] = df['absolute_max'] / (df['rms'] ** 2)  # Clearance factor: 절대 최대값 / RMS
```
